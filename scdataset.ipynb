```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:/sc_data.csv")
df.head()
```

```python
highest_market_capital_country= df.groupby('country')['market_cap_04_08_2025'].sum().sort_values(ascending = False)
highest_market_capital_country = pd.DataFrame(highest_market_capital_country)
```

```python
plt.figure(figsize = (12,8))
highest_market_capital_country.plot(kind = 'bar',color = 'black')
plt.title('Highest_Capital_Countries')
plt.xlabel('Countries')
plt.ylabel('Market_Capital')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

```python
highest_market_capital_country.to_csv('highestmarketcapitalcountry.csv',index = True)
```

```python
highest_capital_companies = df.groupby('name')['market_cap_04_08_2025'].sum().sort_values(ascending = False)
```

```python
top10highestcapitalcompanies = highest_capital_companies.head(10)
top10highestcapitalcompanies = pd.DataFrame(top10highestcapitalcompanies)
```

```python
top10highestcapitalcompanies.to_csv('top10highestcapitalcompanies.csv',index = True)
```

```python
plt.figure(figsize = (12,8))
top10highestcapitalcompanies.plot(kind = 'bar',color = 'black')
plt.title('Highest_Capital_Companies')
plt.xlabel('Companyname')
plt.ylabel('MarketCapital')
plt.show()
```

```python
plt.scatter(df['market_cap_04_08_2025'],df['price_04_08_2025'])
plt.title('Price vs Market Capital')
plt.xlabel('Market_Capital')
plt.ylabel('Price')

```
